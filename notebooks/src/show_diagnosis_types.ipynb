Importações e Setup

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura
df = pd.read_csv('data/raw/metadata/dermoscopic_100_000.csv', low_memory=False)

# 2. Empilha as 5 colunas de diagnóstico numa única série
diag_cols = ['diagnosis_1', 'diagnosis_2', 'diagnosis_3', 'diagnosis_4', 'diagnosis_5']
all_diags = df[diag_cols] \
    .melt(value_name='diagnosis') \
    ['diagnosis'] \
    .dropna()

# 3. Conta frequência de cada diagnóstico
counts = all_diags.value_counts()

# 4. Exibe a contagem
print("Frequência de cada diagnóstico:\n", counts[:20])

# 5. Plota gráfico de barras
plt.figure(figsize=(10,6))
counts.plot(kind='bar')
plt.title('Distribuição de Diagnósticos')
plt.xlabel('Diagnóstico')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
